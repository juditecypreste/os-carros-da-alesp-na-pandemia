```python
# Importação de bibliotecas e módulos
import pandas as pd
import numpy as npb
```

```python
# Me mostre tudo
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Abrindo o df de despesas
despesas = pd.read_csv("despesas_gabinetes.csv", ",", encoding="utf-8")
```

```python
# Vendo meu df
despesas.head()
```

```python
# Abrindo o arquivo com matrículas dos deputados
deputados = pd.read_csv("lista_deputados.csv")
```

```python
# Unindo os dfs
df = pd.merge(despesas, deputados, how='inner', on='Matricula', validate='many_to_many')
df.head()
```

```python
# Conferindo os anos disponíveis
df.Ano.unique()
```

```python
# Selecionando apenas 2020
ano_2020 = df[df['Ano'] == 2020]
ano_2020.head()
```

```python
#Selecionando apenas o primeiro semestre
ano_2020 = ano_2020[ano_2020["Mes"].isin([1,2,3,4,5,6])]
```

```python
# Conferindo qual foi a maior despesa dos deputados naquele ano
ano_2020.groupby('Tipo')['Valor'].sum()
```

```python
# Selecionando apenas 2021
ano_2021 = df[df['Ano'] == 2021]
ano_2021.head()
```

```python
#Selecionando apenas o primeiro semestre
ano_2021 = ano_2021[ano_2021["Mes"].isin([1,2,3,4,5,6])]
```

```python
# Conferindo qual foi a maior despesa dos deputados naquele ano
ano_2021.groupby('Tipo')['Valor'].sum()
```

```python
# Selecionando apenas os gastos com locação de veículos
ano_2021 = ano_2021[ano_2021['Tipo'] == 'O - LOCAÇÃO DE VEÍCULO']
```

```python
# Conferindo quem gastou mais com locações
ano_2021.groupby('Deputado')['Valor'].sum().sort_values(ascending=False)
```

```python
# Quantos deputados locaram veículos?
ano_2021['Deputado'].nunique()
```

```python
# Qual foi o fornecedor que mais recebeu dinheiro naquele ano
ano_2021.groupby('Fornecedor')['Valor'].sum()
```

```python
# Conferindo todas as locações feitas com a Nevada
ano_2021 = ano_2021[ano_2021['Fornecedor'] == 'NEVADA RENT A CAR LTDA']
```
