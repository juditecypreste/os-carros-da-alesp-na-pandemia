```python
# Importação de bibliotecas e módulos
import pandas as pd
import numpy as np
import xml.etree.cElementTree as et
```

```python
# Abrindo o arquivo baixado
tree = et.parse('deputados.xml')
root = tree.getroot()
```

```python
# Criando listas 
Deputado = []
Matrícula = []
```

```python
# Registrando os deputados e suas matrículas
for deputado in root.iter('NomeParlamentar'):
    print(deputado.text)
    Deputado.append(deputado.text)

for matrícula in root.iter('Matricula'):
    print(matrícula.text)
    Matrícula.append(matrícula.text)
```

```python
# Transformando isso em um df
df = pd.DataFrame(list(zip(Deputado, Matrícula)), 
                       columns=['Deputado', 'Matricula'])
```

```python
# Vendo meu df
df.head()
```

```python
# Apagando duplicadas
df.drop_duplicates()
```

```python
# Salvando o df com o nome dos deputados e suas matrículas
df.to_csv("lista_deputados.csv")
```
